```python
# Which was the most ordered item? and How many items were ordered?

item=df.groupby('item_name')

df1=item.sum()
df2=df1.sort_values('quantity',ascending=False)
df2.head(1)
```

```python
# What was the most ordered item in the choice_description column?
 
choice=df.groupby('choice_description')

df3=choice.sum()
df4=df3.sort_values('quantity',ascending=False)
df4.head(1)    
```

```python
# Turn the item price into a float


def getprice(strprice):
    return strprice[1:]


df['new_itemprice']=df['item_price'].apply(getprice)
df5=df.astype({"new_itemprice":float})
df5
df5.dtypes
```

```python
# print a data frame with only two columns item_name and item_price


df[['item_name','item_price']]
```

```python
# delete the duplicates in item_name and quantity


df.drop_duplicates(['item_name','quantity'])
```

```python
# select only the products with quantity equals to 1


df[df['quantity']==1]
```

```python
# sort item from the most to less expensive




exp=df5.groupby('item_name')

df6=exp.sum()
df7=df6.sort_values('new_itemprice',ascending=False)
df7.head() 
```

```python
# What was the quantity of the most expensive item ordered?


df8=df7[['quantity','new_itemprice']]
df8.head(1)
```
